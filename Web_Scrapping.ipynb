```python
# Web Scrapping with Python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Import libraries for url manipulation
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
url="https://www.nurseryrhymes.org/nursery-rhymes.html"
html=urlopen(url)
```

```python
soup=BeautifulSoup(html,'lxml')
type(soup)
```

```python
# Get the title
title=soup.title
print(title)
```

```python
# Check the text on webpages
text=soup.get_text()
print(soup.text)
```

```python
all_links=soup.find_all('td')
all_links
```

```python
all_links=soup.find_all('a')
#all_links1=all_links.get('a')
links=[]
for link in all_links:
  links.append(link.get("href"))
 # print(link.get("href"))

links
```

```python
len(links)
links_new=links[24:]
print(len(links),len(links_new))
```

```python
# Check one song
url="https://www.nurseryrhymes.org/a-sailor-went-to-sea.html"
html=urlopen(url)
soup=BeautifulSoup(html,'lxml')
type(soup)
```

```python
all_rhymes=str(soup.find('em'))
x = all_rhymes.replace("<em>", "")
x = all_rhymes.replace("</em>", "")
x = all_rhymes.replace("<br/>", "")
#for link in all_links:
#print(link.get("href"))
```

```python
for i in range(len(links_new)):
  url="https://www.nurseryrhymes.org/"+ links_new[i] 
  print(url)
  html=urlopen(url)
  soup=BeautifulSoup(html,'lxml')
  all_rhymes=str(soup.find('em'))
  x = all_rhymes.translate({ord('<em>'): None}) 
  x = all_rhymes.replace("</em>", "")
  x = all_rhymes.replace("<br/>", "")
  f = open("\Downloads", 'w')
  file=links_new[i]+'.txt'
  with open(file, 'w') as f:
    f.writelines(x)
```
